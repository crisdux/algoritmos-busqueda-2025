```python
def busqueda_binaria(lista: list, objetivo: int) -> int | None:
    if not lista:
        return None
    
    lista_ordenada = sorted(lista)
    index_inicio = 0
    index_final = len(lista_ordenada) - 1

    while index_final >= index_inicio:
        index_medio = (index_inicio + index_final) // 2
        
        if lista_ordenada[index_medio] > objetivo:
            index_final = index_medio - 1
        elif lista_ordenada[index_medio] < objetivo:
            index_inicio = index_medio + 1
        else:
            return index_medio
    
    return None

print(busqueda_binaria([2, 4, 6, 8, 10, 12, 14, 16], 6)) 
```

```python
import matplotlib.pyplot as plt
import time
import random

def medir_tiempo_busqueda_binaria(n):
    lista = sorted(random.sample(range(1, n*10), n))
    objetivo = random.choice(lista)
    
    inicio = time.time()
    busqueda_binaria(lista, objetivo)
    fin = time.time()
    
    return fin - inicio

# Tamaños de entrada
tamanos = [10, 100, 1000, 10000, 100000, 1000000]
tiempos = []

# Medir tiempos para cada tamaño de entrada
for n in tamanos:
    tiempo = medir_tiempo_busqueda_binaria(n)
    tiempos.append(tiempo)

# Graficar los resultados
plt.plot(tamanos, tiempos, marker='o')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Tamaño de la lista (n)')
plt.ylabel('Tiempo de ejecución (segundos)')
plt.title('Complejidad de la Búsqueda Binaria O(log n)')
plt.grid(True)
plt.show()
```
