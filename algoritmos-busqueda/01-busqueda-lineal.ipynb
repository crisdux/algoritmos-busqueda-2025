# Busqueda Secuencial
Tambien denominado busqueda líneal o busqueda por fuerza bruta. 
Consiste en buscar un elemento `x` recorriendo toda la estructura de datos donde lo busquemos hasta encontrarlo caso contrario regresamos `None`.

Veamos un ejemplo:
Dada la lista `numeros = [58,36,41,78,96,125,78,30,2,1,28]` implementar busqueda lineal para encontrar el valor `x=1`.

```python
def busqueda_secuencial(lista:list, objetivo:int):
    for item in lista:
        if item == objetivo:
            return lista.index(item)
    return None

print(busqueda_secuencial([58,36,41,78,96,125,78,30,2,1,28], 1)) 
## regresa 9, lo que significa que el item se encontro en la posición 9 de la lista
```

## Probamos el algoritmo con notación Big(O)

```python
import time
import matplotlib.pyplot as plt

# Configuración del experimento
tamanios = [1000, 5000, 10000, 20000, 50000, 100000]  # Tamaños de lista a probar
tiempos = []  # Almacenará los tiempos de ejecución

for n in tamanios:
    lista = list(range(n+1))  # Lista ordenada [0, 1, 2, ..., n+1]
    valor = -1  # Valor que NO está en la lista (para forzar el peor caso)
    
    # Medición del tiempo
    inicio = time.perf_counter()
    busqueda_secuencial(lista, valor)
    fin = time.perf_counter()
    
    tiempos.append(fin - inicio)

# Generación del gráfico
plt.figure(figsize=(10, 6))
plt.plot(tamanios, tiempos, 'o-', label="Tiempo experimental", color="blue")
plt.xlabel("Tamaño de la lista (n)", fontsize=12)
plt.ylabel("Tiempo de ejecución (segundos)", fontsize=12)
plt.title("Complejidad de la Búsqueda Lineal: O(n)", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

## Notación Big(O)
Podemos notar por la grafica que a medida que crece la lista, la busqueda tiene a ser lineal, dicho de otra manera, ele peor de los casos donde el item `x` a encontrar no estuviera en la lista, el algoritmo de busqueda líneal tendra que recorrer toda la lista. Esto consume tiempo de ejecución y por ende recursos computacionales.

Se concluye que la complejidad de la busqueda es de Líneal O(n), a muestras mas grandes mayor será el tiempo de ejecución.

Dicho de otra forma, si tengo una lista con 10000 registros y el item que quiero buscar se encuentra en la última posición (el peor de los casos posibles) entonces tendria que la complejidad del algotirmo lineal seria de O(10000), osea que tuvimos que dar 10000 iteraciones para encontrar nuestro elemento. 



En nuestra escala de complejidades algoritmicas O(n) se encuentra en "Fear" lo que se puede traducir en español como un mas o menos, no esta mal pero puede mejorarse. 

## Ventajas y Desventajas

| Pros                                      | Contras                                             |
|-------------------------------------------|-----------------------------------------------------|
| Sencillo y fácil de implementar.          | No es adecuado para grandes conjuntos de datos debido al largo tiempo de ejecución. |
| Funciona en matrices sin ordenar.         | Tiene una complejidad temporal O(n) en el peor de los casos. |
| Adecuado para conjuntos de datos pequeños. |                                                     |
